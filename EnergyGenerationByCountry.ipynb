```python
countries = ['World',
             'Asia Pacific',
             'North America',
             'South & Central America',
             'Africa',
             'Europe (other)',
             'Europe',
             'CIS',
             'Other South America',
             'Middle East',
             'Other CIS']

worldFiltered = data.copy()

for country in countries:
  worldFiltered = worldFiltered[worldFiltered['country']!= country ]
```
